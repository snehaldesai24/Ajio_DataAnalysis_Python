```python
#To check duplicate in dataframe
orders.duplicated().sum()
transactions.duplicated().sum()
ratings.duplicated().sum()
customers.duplicated().sum()
delivery.duplicated().sum()
```

```python
#if there are any duplicated values then use

orders = orders.drop_duplicates()
orders
```

```python
#To check for null values in dataframe
orders.isnull().sum()
```

```python
#If there are null values
#drop nulls - (loss of data)
# fill nulls - (constant, mean, mode, median, ffill, bfill)
#if numberic col, use Median
#if textual col, use Mode
#backward fill and forward fill
#assuming quantity has null values
orders["Qty"] = orders["Qty"].fillna(orders["Qty"].median()) #mean() or Mod
orders["Qty"] = orders["Qty"].ffill()  #bfill
```

```python
orders.head()
```

```python
#join tables
merge_cust_ord = pd.merge(right=customers, left=orders, how='inner', on='C_ID')
merge_cust_ord
```

```python
gb=merge_cust_ord.groupby('State').agg({'Or_ID':'count'})
gb_sort=gb.sort_values(by='Or_ID',ascending=False)
gb_sort
```

```python
sns.barplot(x=gb_sort.index,y="Or_ID",data=gb_sort,hue=gb_sort.index,palette="viridis")
plt.xticks(rotation=90)
plt.title("State Wise Distribution of Orders")
plt.show()
```

```python
gb_gender=merge_cust_ord.groupby('Gender').agg({'Or_ID':'count'})
gb_sort1=gb_gender.sort_values(by='Or_ID',ascending=False)
gb_sort1
```

```python
sns.barplot(x=gb_sort1.index,y="Or_ID",data=gb_sort1,hue=gb_sort1.index,palette="viridis")
plt.title("Gender Wise Distribution of Orders")
plt.show()
```

```python
gb_age=merge_cust_ord.groupby('Age').agg({'Or_ID':'count'})
gb_sort2=gb_age.sort_values(by='Or_ID',ascending=False)
gb_sort2=gb_sort2.head(5)
gb_sort2
```

```python
sns.barplot(x=gb_sort2.index,y="Or_ID",data=gb_sort2,order=gb_sort2.index,hue=gb_sort2.index,palette="plasma")
plt.title("Top 5 Age Groups with the Higher Order Volume")
plt.show()
```

```python
orders['month'] = pd.to_datetime(orders['Order_Date']).dt.to_period('M')
monthly_orders = orders.groupby('month')['Or_ID'].nunique().reset_index()
monthly_orders['month'] = monthly_orders['month'].astype(str)
sns.lineplot(data=monthly_orders, x='month', y='Or_ID', marker='o')
plt.title("Monthly Order Trend")
plt.xlabel("Month")
plt.ylabel("Total Orders")
plt.xticks(rotation=90)
plt.show()
```

```python
#Data Combining
data=pd.merge(left=orders,right= customers,on="C_ID",how= "inner")
data=pd.merge(left=orders,right= delivery,on="DP_ID",how= "inner")
data=pd.merge(left=orders,right= products,on="P_ID",how= "inner")
data=pd.merge(left=orders,right= ratings,on="Or_ID",how= "inner")
#data=pd.merge(left=orders,right= returns,on="Or_ID",how= "inner")
data=pd.merge(left=orders,right= transactions,on="Or_ID",how= "inner")
data
```

```python
sales_by_date = data.groupby('Order_Date')['Qty'].agg(['sum', 'mean']).reset_index()
sales_by_date.head(2)
```

```python
#Sales Volume by Date
plt.figure(figsize=(10, 4))
plt.plot(sales_by_date['Order_Date'], sales_by_date['sum'], marker='o', label='Total Qty')
plt.plot(sales_by_date['Order_Date'], sales_by_date['mean'], marker='x', label='Avg Qty')
plt.title("Sales Volume by Date")
plt.xlabel("Order Date")
plt.ylabel("Quantity")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Total & Average Quantity by Product
sales_by_product = data.groupby('P_ID')['Qty'].agg(['sum', 'mean']).sort_values(by='sum', ascending=False).head(10).reset_index()
sales_by_product.head(2)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(sales_by_product['P_ID'], sales_by_product['sum'], alpha=0.6, label='Total Qty')
plt.plot(sales_by_product['P_ID'], sales_by_product['mean'], color='red', marker='o', label='Avg Qty')
plt.title("Top 10 Products by Sales Volume")
plt.xlabel("Product ID")
plt.ylabel("Quantity")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Coupon usage behavior by customer
#1. Normalize "Coupon" column (treat "No Coupon" as NaN for analysis)
data['Coupon_Used'] = data['Coupon'].apply(lambda x: x if x != 'No Coupon' else None)
#2. Coupon Usage Count by Customer
coupon_usage = data[data['Coupon_Used'].notna()].groupby('C_ID')['Coupon_Used'].count().sort_values(ascending=False).head(10)
coupon_usage.head(2)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=coupon_usage.index, y=coupon_usage.values, palette='magma')
plt.title("Top 10 Customers by Coupon Usage")
plt.xlabel("Customer ID")
plt.ylabel("Coupons Used")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Coupon Usage Rate by Customer (% of orders with coupons)
total_orders = data.groupby('C_ID')['Coupon'].count()
coupon_orders = data[data['Coupon_Used'].notna()].groupby('C_ID')['Coupon_Used'].count()
coupon_usage_rate = (coupon_orders / total_orders).fillna(0).sort_values(ascending=False).head(10)
print("total_orders ",total_orders.head(2))
print("couon_orders ",coupon_orders.head(2))
print("coupon_usages_rate ",coupon_usage_rate.head(2))
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=coupon_usage_rate.index, y=coupon_usage_rate.values * 100, palette='coolwarm')
plt.title("Top 10 Customers by Coupon Usage Rate (%)")
plt.xlabel("Customer ID")
plt.ylabel("Coupon Usage Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
